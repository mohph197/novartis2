```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
vol = pd.read_csv('data/test/df_volume_test.csv')
gxs = pd.read_csv('data/test/df_generics_test.csv')
info = pd.read_csv('data/test/df_medicine_info_test.csv')
```

```python
df = vol.merge(
    gxs,
    on=['country', 'brand_name', 'months_postgx'],
    how='left',
    validate='one_to_one'
)
df = df.merge(
    info,
    on=['country', 'brand_name'],
    how='left',
    validate='many_to_one'
)
```

```python
df.head()
```

```python
df.shape
```

```python
full_grid = (
    df.groupby(['country','brand_name'])['months_postgx']
      .apply(lambda s: pd.Series(range(s.min(), s.max()+1)))
      .reset_index()
      .rename(columns={'level_2': 'dummy', 0: 'months_postgx'})
      .drop(columns='dummy')
)

df = full_grid.merge(df, on=['country','brand_name','months_postgx'], how='left')
```
